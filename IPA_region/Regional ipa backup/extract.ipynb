```python
# read docx file
%pip install python-docx
import docx
doc = docx.Document('doc.docx')
print(len(doc.paragraphs))
```

```python
all_text = ""
for para in doc.paragraphs:
    all_text += para.text
print(all_text)
```

```python
numbers=["১","২","৩","৪","৫","৬","৭","৮","৯","০"]
dari="।"
```

```python
for i in range(len(all_text)):
    chr=all_text[i]
    if chr == dari and all_text[i-1] in numbers:
        all_text=all_text[:i+1]+"#"+all_text[i+1:]
```

```python
all_text=all_text.replace("অনুবাদ","@")
all_text=all_text.replace("মর্মার্থ","&")
all_text=all_text.replace("English","$")
all_text=all_text.replace("ENGLISH","$")
print(all_text)
```

```python
start=False
end=False
sentences=[]
sentence=""
print(all_text)
for chr in all_text:
    if chr == "#":
        start=True
        continue
    if chr in ["@","&","$"]:    
        end=True
    
    if start and not end:
        sentence+=chr
    
    elif end:
        if(len(sentence)>0):
            sentences.append(sentence)
        sentence=""
        start=False
        end=False
    
    

print(sentences)

      
```
